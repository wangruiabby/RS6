```python
import csv
import matplotlib.pyplot as plt
from wordcloud import WordCloud
# 数据加载
csv_file=open('./Market_Basket_Optimisation.csv')    #打开csv文件
basket_data = csv.reader(csv_file)   #逐行读取csv文件
```

```python
basket_data_list = []

for line in basket_data:
    for item in line:
        if item is None:
            continue
        basket_data_list.append(item)
#         basket_data_list.append(item.replace(" ", "_"))
basket_data_list1 = " ".join(basket_data_list)
```

```python
my_wordcloud = WordCloud().generate(basket_data_list1)

plt.imshow(my_wordcloud)

plt.axis("off")

plt.show()
```

```python
count_dict = {}
for str in basket_data_list:
    if str in count_dict.keys():
        count_dict[str] = count_dict[str] + 1
    else:
        count_dict[str] = 1
#按照词频从高到低排列
count_list=sorted(count_dict.items(),key=lambda x:x[1],reverse=True)

print("TOP 10 的商品分别是")
print(count_list[0:10])
```
